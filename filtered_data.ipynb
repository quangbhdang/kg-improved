```python
from huggingface_hub import hf_hub_download
import pandas as pd
```

```python
file_path = hf_hub_download(
    repo_id="Zihan1004/FNSPID",
    filename="Stock_news/nasdaq_exteral_data.csv",
    repo_type="dataset"
)
```

```python
df = pd.read_csv(
    file_path,
    dtype=str,
    low_memory=False,
    encoding='utf-8',
    on_bad_lines='skip'  # Skip problematic lines
)
```

```python
df.head(5)
```

```python
# Convert Data column to datetime
df['Date'] = pd.to_datetime(df['Date'])
df_filtered = df[df['Date'].dt.year == 2023]
print(f"Original data shape: {df.shape}")
print(f"Filtered data shape: {df_filtered.shape}")
```

```python
# Count sample data for each month in 2023
monthly_counts = df_filtered.groupby(df_filtered['Date'].dt.to_period('M')).size()
print(monthly_counts)
```

```python
# Filter and save data for Dec 2023
df_of_interest = df_filtered[df_filtered['Date'].dt.to_period('M') == '2023-12']
print(df_of_interest)
output_file = 'nasdaq_dec2023.csv'
df_of_interest.to_csv(output_file, index=False)
```
